```python
import pandas as pd
import json

# Define file paths
business_file_path = 'yelp_academic_dataset_business.json'
review_file_path = 'yelp_academic_dataset_review.json'
user_file_path = 'yelp_academic_dataset_user.json'
```

```python
# Read and process the first 1000 samples from business.json
business_df = pd.read_json(business_file_path, lines=True, nrows=5000)

# Read and process the first 1000 samples from review.json
review_df = pd.read_json(review_file_path, lines=True, nrows=5000)

# Read and process the first 1000 samples from user.json
user_df = pd.read_json(user_file_path, lines=True, nrows=5000)
```

```python
# Merge DataFrames
merged_df = pd.merge(review_df, business_df, on='business_id', how='inner')
merged_df = pd.merge(merged_df, user_df, on='user_id', how='inner')
```

```python
# # Drop irrelevant columns
# irrelevant_columns = ['address', 'city', 'state', 'hours']
# merged_df = merged_df.drop(irrelevant_columns, axis=1)
```

```python
# Handling missing values
merged_df = merged_df.dropna()  # Drop rows with any missing values

# Convert data types if needed (e.g., date)
merged_df['date'] = pd.to_datetime(merged_df['date'])
user_df['yelping_since'] = pd.to_datetime(user_df['yelping_since'])

# Display the first few rows of the merged DataFrame
merged_df
```

```python
csv_output_path = 'merged_2K_data.csv'
merged_df.to_csv(csv_output_path, index=False)
```
